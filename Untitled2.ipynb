```python
#1. Dataset Preparation
```

```python
import pandas as pd
```

```python
# Load the dataset

# list of possible encodings to try
encodings = ['utf-8', 'latin1', 'ISO-8859-1', 'cp1252']

file_path = 'books.csv'   #change this to the path of the csv file

# attempt to read the csv files with different encoding:
for encoding in encodings:
   try:
      df = pd.read_csv(file_path, encoding=encoding)
      print(f"File successfully read with encoding: {encoding}") 
      break        #Stop the loop if successful
   except UnicodeDecodeError:
      print(f"Failed to read with encoding: {encoding}") 
      continue    #Try the next encoding

#If the Loop completes without success, df will not be defined 
if 'df' in locals():
   print("CSV file has been successfully loaded.")
else:
    print("All encoding attempts failed. Unable to read the CSV file.")
```

```python
## Clustering Implementation
```

```python
#A. K-Means Clustering
```

```python
column_names = df.columns
print(column_names)
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Preprocess the data (standardization)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df[['Unnamed: 0', 'Price', 'Price_After_Tax', 'Tax_amount', 'Avilability', 'Number_of_reviews', 'Stars']])

# Implementing K-Means Clustering
kmeans = KMeans(n_clusters=5, random_state=42)
kmeans.fit(X_scaled)

# Adding cluster labels to the original dataframe
df['Cluster_KMeans'] = kmeans.labels_
```

```python
#B. Hierarchical Clustering
```

```python
from sklearn.cluster import AgglomerativeClustering

# Implementing Hierarchical Clustering
hierarchical = AgglomerativeClustering(n_clusters=5)
df['Cluster_Hierarchical'] = hierarchical.fit_predict(X_scaled)
```

```python
#C. DBSCAN 
```

```python
from sklearn.cluster import DBSCAN

# Implementing DBSCAN
dbscan = DBSCAN(eps=0.3, min_samples=10)
df['Cluster_DBSCAN'] = dbscan.fit_predict(X_scaled) 
```

```python
#3. Model Evaluation
```

```python
from sklearn.metrics import silhouette_score, davies_bouldin_score

def evaluate_clustering(X, labels):
    """ Evaluate clustering using silhouette score and Davies-Bouldin index """
    if len(set(labels)) > 1:
        silhouette = silhouette_score(X, labels)
        dbi = davies_bouldin_score(X, labels)
    else:
        silhouette = 'Not applicable (only one cluster)'
        dbi = 'Not applicable (only one cluster)'
    return silhouette, dbi

# Evaluate K-Means
silhouette_kmeans, dbi_kmeans = evaluate_clustering(X_scaled, df['Cluster_KMeans'])

# Evaluate Hierarchical Clustering
silhouette_hierarchical, dbi_hierarchical = evaluate_clustering(X_scaled, df['Cluster_Hierarchical'])

# Evaluate DBSCAN
# Note: DBSCAN might label noise points as -1, so filter them out before evaluation
filtered_labels = df['Cluster_DBSCAN'][df['Cluster_DBSCAN'] != -1]
filtered_X_scaled = X_scaled[df['Cluster_DBSCAN'] != -1]

silhouette_dbscan, dbi_dbscan = evaluate_clustering(filtered_X_scaled, filtered_labels)

# Print the results
print(f"Silhouette Score - KMeans: {silhouette_kmeans}, Hierarchical: {silhouette_hierarchical}, DBSCAN: {silhouette_dbscan}")
print(f"Davies-Bouldin Index - KMeans: {dbi_kmeans}, Hierarchical: {dbi_hierarchical}, DBSCAN: {dbi_dbscan}")
```

```python
#4. Visualization
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# K-Means Clustering Visualization
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df['Price'], y=df['Price_After_Tax'], hue=df['Cluster_KMeans'], palette='viridis')
plt.title('K-Means Clustering')
plt.xlabel('Price')
plt.ylabel('Price After Tax')
plt.legend(title='Cluster')
plt.show()

# Hierarchical Clustering Visualization
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df['Price'], y=df['Price_After_Tax'], hue=df['Cluster_Hierarchical'], palette='viridis')
plt.title('Hierarchical Clustering')
plt.xlabel('Price')
plt.ylabel('Price After Tax')
plt.legend(title='Cluster')
plt.show()

# DBSCAN Clustering Visualization
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df['Price'], y=df['Price_After_Tax'], hue=df['Cluster_DBSCAN'], palette='viridis')
plt.title('DBSCAN Clustering')
plt.xlabel('Price')
plt.ylabel('Price After Tax')
plt.legend(title='Cluster')
plt.show()
```
